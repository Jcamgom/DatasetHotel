# Análisis Univariado

### Se escogieron 5 columnas para el analisis univariado:

lead_time

El tiempo de anticipación de la reserva es crucial para la planificación y gestión de la ocupación hotelera. Un mayor lead time puede permitir una mejor preparación y asignación de recursos, mientras que un lead time corto puede indicar reservas de última hora. Analizar esta variable puede ayudar a identificar patrones en el comportamiento de los clientes y a prever la demanda futura. También es útil para entender si las reservas con mayor anticipación tienen una mayor probabilidad de cancelación.

adr

La tarifa diaria promedio es una métrica clave para evaluar los ingresos y la rentabilidad del hotel. Es fundamental para la estrategia de precios y para entender cómo diferentes factores afectan los ingresos. Analizar esta variable puede revelar cómo las tarifas varían según la temporada, el tipo de cliente, y otros factores. También puede ayudar a identificar oportunidades para ajustar precios y maximizar ingresos.

stays_in_weekend_nights

Las estadías de fin de semana suelen tener un comportamiento diferente al de las estadías entre semana. Los fines de semana pueden atraer a más turistas y viajeros de ocio, lo que puede influir en la ocupación y las tarifas. Analizar esta variable puede ayudar a entender la demanda específica de los fines de semana y a ajustar las estrategias de marketing y precios para maximizar la ocupación y los ingresos durante estos días.

stays_in_week_nights

Las estadías entre semana suelen estar más relacionadas con viajeros de negocios y estancias más largas. Comprender este patrón es esencial para la gestión de la ocupación y la planificación de recursos. Analizar esta variable puede proporcionar insights sobre la demanda entre semana y ayudar a identificar oportunidades para atraer a más clientes durante estos días, optimizando así la ocupación y los ingresos.

is_canceled

La cancelación de reservas es un factor crítico que afecta la ocupación y los ingresos del hotel. Entender las razones y patrones detrás de las cancelaciones puede ayudar a reducir su incidencia y mejorar la estabilidad financiera. Analizar esta variable en relación con otras puede revelar factores que influyen en la cancelación de reservas, permitiendo al hotel implementar políticas y estrategias para minimizar las cancelaciones y mejorar la retención de clientes.


```python
plt.figure(figsize=(10, 6))
sns.histplot(df['lead_time'], bins=50, kde=True)
plt.title('Distribución de Lead Time')
plt.xlabel('Lead Time (días)')
plt.ylabel('Frecuencia')
plt.show()
```

La mayoría de los clientes reservan con una anticipación de entre 0 y 200 días, con algunos valores extremos que llegan hasta 737 días. La distribución muestra una alta variabilidad con una media de 104 días.

```python
#Lead Time
plt.figure(figsize=(10, 6))
sns.histplot(df['lead_time'][df['lead_time'] <= 50], bins=50, kde=True)
plt.title('Distribución de Lead Time (<= 50 días)')
plt.xlabel('Lead Time (días)')
plt.ylabel('Frecuencia')
plt.show()
```

Siendo más especificos en el rango se puede observar como la mayoria de personas reserva con muy poca anticipación entre 1 a 10 dias.

```python
# adr
plt.figure(figsize=(10, 6))
sns.histplot(df['adr'], bins=50, kde=True)
plt.title('Distribución de ADR (Average Daily Rate)')
plt.xlabel('ADR (Tarifa Diaria Promedio)')
plt.ylabel('Frecuencia')
plt.show()
```

La mayoría de las tarifas diarias se encuentran entre 0 y 200 unidades monetarias, con algunos valores extremos que superan los 500. La distribución tiene una media de 101.83 unidades monetarias. Hay un outlier de 5400.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['adr'][df['adr'] < 200], bins=50, kde=True)
plt.title('Distribución de ADR (Average Daily Rate)')
plt.xlabel('ADR (Tarifa Diaria Promedio)')
plt.ylabel('Frecuencia')
plt.show()
```

Siendo más especificos en el rango se puede observar claramente como la tarifa varia entre 0 y 200 .

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['stays_in_weekend_nights'], bins=30, kde=True)
plt.title('Distribución de Noches de Fin de Semana')
plt.xlabel('Noches de Fin de Semana')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['stays_in_weekend_nights'][df['stays_in_weekend_nights'] <= 2], bins=30, kde=True)
plt.title('Distribución de Noches de Fin de Semana (<= 2)')
plt.xlabel('Noches de Fin de Semana')
plt.ylabel('Frecuencia')
plt.show()
```

La mayoría de los huéspedes se quedan 0 noches de fin de semana. Es decir, son huespedes que se quedan solo entre semana .
